```python
# Add the Pandas dependency
import pandas as pd
import datetime
```

```python
# Files to load
file_to_load = "database_clean.csv"

# Read the CSV into a DataFrame
data_df = pd.read_csv(file_to_load)
data_df
```

```python
# Add index and save it as 'patient ID'
data_df.index.name = "patient_id"
data_df = data_df.reset_index()
data_df.head()
```

```python
# Show columns
data_df.columns
```

```python
# Variable for new column names
keymap = {
    'masculino=1': 'gender',
    'estado': 'state',
    'Cod INEGI': 'cod_inegi',
    'municipio': 'municipality',
    'Altura': 'altitude',
    'edad en días': 'age',
    'Peso    (kg)': 'weight_kg',
    'Talla    (cm)': 'height_cm',
    'Apariencia Nutricional (Normal/Desnutrido/ Emaciado/ Sobrepeso)': 'appearance',
    'Prematurez': 'premature',
    'Diagnóstico No cardiaco       (Estructural/ Cromosómica/ Síndrome/ Enfermedad)': 'diagnosis_general',
    'Cx Previas Cardiaca                         (Numero)': 'cx_previous',
    'Diagnóstico Principal': 'diagnosis_main',
    'Fx nacimiento': 'date_birth',
    'fecha de diagnóstico': 'date_diagnosis',
    'fecha de cirugía': 'date_procedure',
    'Cirugía': 'procedure',
    'RACHS': 'rachs',
    'Estancia UCICV                                 (días, con formula)': 'stay_days',
    'muerte': 'expired'   
}
# Risk Adjustment for Congenital Heart Surgery
```

```python
# Change of names for all columns
data_df = data_df.rename(columns=keymap)
data_df.head()
```

```python
# Verify the number of rows in each column
data_df.isnull().sum()
```

```python
# Drop Cod INEGI from df
data_df = data_df.drop(columns='cod_inegi')
```

```python
# Drop "Prematurez" column (incomplete info)
data_df = data_df.drop(columns='premature')
```

```python
# Drop "Prematurez" column (incomplete info)
data_df = data_df.drop(columns='date_diagnosis')
```

```python
# Drop unnecessary columns
data_df = data_df.drop(columns= ['Residuo edad en dias', 'Edad en años', 'Días (edad)', 'Meses (edad)'])
```

```python
# Verify the number of rows in each column
data_df.isnull().sum()
```

```python
# Drop null rows (there is no way to know the values)
data_df = data_df.dropna(subset=['weight_kg', 'height_cm', 'rachs', 'stay_days', 'diagnosis_general'])
data_df.isnull().sum()
```

```python
# Verify the number of rows in the database (must have 1000 rows or more)
data_df.count()
```

```python
# Show type of data
data_df.dtypes
```

```python
# Change data types to the correct format
data_df['gender'] = data_df['gender'].astype(int)
data_df['cx_previous'] = data_df['cx_previous'].astype(int)
data_df['date_birth'] = pd.to_datetime(data_df['date_birth']) 
data_df['date_procedure'] = pd.to_datetime(data_df['date_procedure'])
data_df['rachs'] = data_df['rachs'].astype(float)
data_df['stay_days'] = data_df['stay_days'].astype(float)

data_df.dtypes
```

```python
data_df.to_csv('../../database/data_clean1.csv', index=False)
```
