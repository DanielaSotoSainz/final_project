```python
# Add the Pandas dependency
import pandas as pd
```

```python
# Files to load
file_to_load = "../database/clean2/db_patient_diagnosis_main.csv"

# Read the CSV into a DataFrame
data_df = pd.read_csv(file_to_load)
data_df
```

```python
# Count the number of different diagnosis
count_diagnosis = data_df["diagnosis_main"].value_counts()
count_diagnosis[0:15]
```

```python
index = count_diagnosis[count_diagnosis < 15].index
data_df["diagnosis_main"] = data_df["diagnosis_main"].map( lambda x: "Other" if x in index else x)
data_df.to_csv("../database/diagnosis_main_other.csv", index = False)
data_df
```
